```python
# Importando bibliotecas
# docs -> https://www.crummy.com/software/BeautifulSoup/bs4/doc/

from bs4 import BeautifulSoup
import requests
```

```python
# criando um re

url = 'https://quotes.toscrape.com/'

# criando um request e retornando o codigo de retorno
r = requests.get(url)
status = r.status_code
print(status)
```

```python
# create a soup
html = r.text
soup = BeautifulSoup(html, 'html.parser')

# retorna os valores do html
print(soup.prettify)
```

```python
soup.title
```

```python
soup.title.string
```

```python
# find all the a tags

soup.find_all('a')
```

```python
# retornando todos os links

for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
# encontrando todos os quotes 

quotes = soup.select('.quote')
list_quotes = []
for quote in quotes:
    texts = quote.select('.text')
    authors = quote.select('.author')
    for txt,author in zip(texts,authors):
        list_quotes.append([txt.text,author.text])
```
